```python
import math
while True:
    while True:
        try:
            num1 = int(input("첫 번째 정수: "))
            break
        except ValueError:
            print("잘못된 입력입니다 정수만 입력해주세요.")
    while True:
        try:
            num2 = int(input("두번째 정수: "))
            break
        except ValueError:
            print("잘못된 정수.")
    calc = input("연산자(+, -, *, /, **): ")
    try:
        if calc == "+":
            result = num1 + num2
        elif calc == "-":
            result = num1 - num2
        elif calc == "*":
            result = num1 * num2
        elif calc == "/":
            result = num1 / num2
        elif calc == "**":
            result = num1 ** num2
        else:
            raise ValueError("지원하지 않는 연산자.")
        print(f"결과: {result}")
    except ZeroDivisionError:
        print("0으로 나눌 수 없음.")
    except ValueError as e:
        print(e)
    again = input("계속 계산? (네/아니요): ")
    if again != '네':
        print("종료.")
        break
```
